```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## 1. Line Plot (plt.plot)
Use When:

You want to show trends over time

Data has a continuous sequence (like dates, months, years)

Comparing patterns or progression

Example:

Monthly sales, stock prices, temperature over days

```python

#Basics
x = [1, 2, 3, 4, 5]
y = [10, 15, 13, 18, 16]


plt.plot(x,y) 
plt.title("Line Plot")
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
plt.grid(True)
plt.show()
```

```python
x = ['Mon','Tue','Wed','Thur','Fri','Sat','Sun']
y = [220,110,100,340,120,240,111]

plt.plot(x,y)
plt.xlabel("Days Of The Week")
plt.ylabel('Sales Per Day')
plt.title('Bakery Sales This Week!')
plt.show()
```

```python
months = ['JAN','FEB','MAR','APR']
sales =[1500,2000,1000,4000]

plt.plot(months,sales, color = 'red', linestyle ='--', linewidth = 2, marker = 'o',markerfacecolor ='black',label = '2025 four months sales')
plt.title('2025 Montly Sales')
plt.xlabel('Months')
plt.ylabel('Sales Per Month')
plt.legend(fontsize = '12')
plt.xlim(0,3)
plt.ylim(0,4000)
plt.xticks(ticks=months, labels=['M1', 'M2', 'M3', 'M4'])
plt.grid(linestyle = ':',color = 'gray')
plt.show()
```

## 2. Bar Chart (plt.bar)
Use When:

You're comparing quantities between categories

Data is discrete (not continuous)

You want to rank or highlight differences

Example:

Sales by product category, number of customers by region

```python
product = ['Pen','Headphones','Table','Toy']
sales =[500,2500,2000,1000]

plt.bar(product,sales, color='blue',label = 'Weekly Sales')
plt.xlabel('Products')
plt.ylabel('Sales Price')
plt.legend()
plt.show()
```

## 3. Pie Chart (plt.pie)

Use When:

You want to show parts of a whole

There are less than 6 categories (avoid clutter)

You're visualizing percentage breakdowns

Avoid if:

Too many slices (hard to read)

Need to compare actual values

```python
#revenue by country
countries = ['Pakistan','India','USA','UAE']
revenue = [100000,400000,550000,110000]
explode = [0, 0, 0.1, 0]
plt.pie(revenue, labels = countries, colors=['gold','skyblue','coral','lightgreen'],autopct='%1.1f%%',shadow=True, startangle=90, explode = explode)
plt.title('Revenue By Country')
plt.legend(title="Countries", loc='upper left' )
plt.axis('equal')  # Keep the pie chart circular

plt.show()
```

## 4. Histogram (plt.hist)
Use When:

You want to show distribution of a numeric variable

You're looking for skewness, spread, frequency

Example:

Distribution of exam scores, income brackets

```python

# Sample student scores
scores = [45, 67, 89, 56, 74, 90, 55, 60, 68, 72, 88, 95, 49, 52, 66, 80, 78, 84, 92, 59]

# Plot histogram
plt.hist(scores, bins=5, color='skyblue', edgecolor='black')

plt.title('Distribution of Student Scores')
plt.xlabel('Score Range')
plt.ylabel('Number of Students')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

## 5. Scatter Plot (plt.scatter)
Use When:

You're exploring relationships between two variables

You want to see correlation, clusters, or outliers

X and Y are both continuous

Example:

Height vs weight, age vs income

```python
studies_hours = [1,2,3,4,5]
exam_scores = [50,55,65,70,80]

plt.scatter(studies_hours,exam_scores, color = 'green',marker = 'o')
plt.title('Relationship Between Study Hours And Exam Score')
plt.xlabel('STUDY HOURS')
plt.ylabel('EXAM SCORE')
plt.grid(True)
#plt.legend()
plt.show()
```

```python
plt.scatter([1,2,3],[50,57,65], color = 'Blue', label ="Class A")
plt.scatter([1,2,3],[45,50,61], color = 'Orange', label ="Class B")
plt.title(' Comparison Of Two Classes')
plt.xlabel('Study hours')
plt.ylabel('Exam Score')
plt.legend()
plt.grid(True)
plt.show()
```

## What is subplot in Matplotlib?

In Matplotlib (Python), subplot is used to create multiple plots in one figure.
It allows you to organize your visualizations in a grid layout (like 2x2, 1x3, etc.) so you can compare them easily.
Common Use Cases:

Scientific experiments: Plot temperature, pressure, and volume separately.

Stock market data: Compare stock trends of multiple companies.

Machine learning: Train vs validation accuracy/loss.

Data summaries: Histogram + Boxplot of the same variable.

```python
x = [1, 2, 3, 4, 5]
y1 = [10, 20, 25, 30, 40]
y2 = [5, 15, 20, 25, 35]

#Create 1 row and 2 columns of subplots
plt.subplot(1, 2, 1)
plt.plot(x, y1)
plt.title("Graph 1")

plt.subplot(1, 2, 2)
plt.bar(x, y2)
plt.title("Graph 2")

plt.tight_layout()
plt.savefig('Charts.jpg', dpi = 300, bbox_inches = 'tight')
plt.show()
```

## in another way(preferred)

```python
fig, axs = plt.subplots(1, 2)

x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 25, 30]
y2 = [5, 7, 12, 18, 25]

axs[0].plot(x,y, color='red')
axs[0].set_title("Line Plot")

axs[1].bar(x,y2, color="green")
axs[1].set_title("Bar Chart")

fig.suptitle("Comparison Of Line And Bar Chart")
plt.tight_layout()
plt.show()
```
