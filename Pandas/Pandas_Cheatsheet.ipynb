## Interpolation in Pandas


```python
#We use interpolation in Pandas in one line to quickly and efficiently fill missing values (NaN) with estimated values,
#making the data usable for analysis or modeling.

data = {
    'A': [1, 2, None, 4, 5],
    'B': [5, None, None, 8, 10]
}
df3 = pd.DataFrame(data)

print(f"Before Interpolation:\n{df3}")

print(f"\nAfter Interpolation:Only colunmn 'B' Values\n")
df3["B"] = df3["B"].interpolate()
print(df3)

print("\ninterpolate the whole DataFrame at once:\n")
df3 = df3.interpolate()
df3
```

## Merging and Joining

### Merging

```python
df_customers = pd.DataFrame({
    "CiD":[1,2,3],
    "Name":["Uzair","Zaid","Zain"]
})

df_orders =pd.DataFrame( {
    "CiD":[1,2,4],
    "Amount":[2000,3000,35000]
})

#merge
#df.merge(df1,df2, how = "inner", on = "CiD" )

#Different how options:
#how	Description
#inner	Keep only matching keys in both DataFrames
#left	Keep all keys from  column1 match from column2 (NaN if no match)
#right	Keep all keys from column1 match from column2
#outer	Keep all keys from both DataFrames, fill NaN where no match

merged = pd.merge(df_customers,df_orders, how= 'outer', on = "CiD" )
merged
```

## Concatenation

### What is concatenation?

Concatenation is different from merging/joining. Instead of combining based on keys, it stacks DataFrames either:

Vertically (one below the other, adding rows)

Horizontally (side by side, adding columns)

```python
# DataFrame 1
df_1 = pd.DataFrame({
    'Name': ['Ali', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})

# DataFrame 2
df_2 = pd.DataFrame({
    'Name': ['David', 'Eva'],
    'Age': [40, 45]
})
# DataFrame 3 (different columns)
df_3 = pd.DataFrame({
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston', 'Phoenix']
})

vertical_concat = pd.concat([df_1, df_2], ignore_index=True)
print(vertical_concat)
print()
horizontal_concat = pd.concat([df_1, df_3], axis=1)
print(horizontal_concat)
```
